# Работа с числами с плавающей точкой в 32-битном NASM: основные положения

Работа с числами с плавающей точкой в 32-битном ассемблере NASM требует глубокого понимания нескольких ключевых моментов. Они включают 
- представление чисел с плавающей точкой в памяти,
- использование специализированных регистров,
- набор инструкций для выполнения арифметических операций.

Рассмотрим каждый из этих аспектов подробно.

## 1. Представление чисел с плавающей точкой в памяти

Числа с плавающей точкой в компьютерах представляются в соответствии со стандартом IEEE 754. Этот стандарт определяет формат хранения чисел, который включает три основных компонента:
- **Знак (Sign)** — определяет, является ли число положительным или отрицательным (1 бит).
- **Экспонента (Exponent)** — определяет порядок числа (8 бит для одинарной точности, 11 бит для двойной).
- **Мантисса (Mantissa)** — определяет значимые цифры числа (23 бита для одинарной точности, 52 бита для двойной).

В 32-битном NASM для работы с такими числами необходимо понимать, как они хранятся в памяти и как их можно загружать в регистры.


## 2. Регистры для работы с числами с плавающей точкой

В 32-битной архитектуре x86 для работы с числами с плавающей точкой используются 
- **регистры FPU (Floating-Point Unit)** 
- **регистры SSE (Streaming SIMD Extensions)**.

### Регистры FPU:
- FPU предоставляет 8 регистров: `ST0`, `ST1`, ..., `ST7`.
- Эти регистры работают как стек (LIFO — Last In, First Out).
- Регистры FPU поддерживают операции с числами одинарной, двойной и расширенной точности.

### Регистры SSE:
- SSE предоставляет 8 регистров: `XMM0`, `XMM1`, ..., `XMM7`.
- Эти регистры поддерживают операции с числами одинарной и двойной точности, а также векторные операции.
- Регистры SSE более современные и эффективные для работы с числами с плавающей точкой, чем FPU.


## 3. Инструкции для работы с числами с плавающей точкой

NASM поддерживает набор инструкций для работы с числами с плавающей точкой, которые можно разделить на несколько категорий:

### Инструкции FPU:
- **Загрузка и сохранение**:
  - `FLD` — загрузить число с плавающей точкой в стек FPU.
  - `FST` — сохранить число из стека FPU в память.
- **Арифметические операции**:
  - `FADD` — сложение.
  - `FSUB` — вычитание.
  - `FMUL` — умножение.
  - `FDIV` — деление.
- **Сравнение**:
  - `FCOM` — сравнить два числа с плавающей точкой.
- **Управление стеком**:
  - `FXCH` — обменять значения в регистрах FPU.

### Инструкции SSE:
- **Загрузка и сохранение**:
  - `MOVSS` — загрузить или сохранить число одинарной точности.
  - `MOVSD` — загрузить или сохранить число двойной точности.
- **Арифметические операции**:
  - `ADDSS`, `ADDSD` — сложение.
  - `SUBSS`, `SUBSD` — вычитание.
  - `MULSS`, `MULSD` — умножение.
  - `DIVSS`, `DIVSD` — деление.
- **Сравнение**:
  - `COMISS`, `COMISD` — сравнить два числа с плавающей точкой.


## 4. Важные замечания

- **Точность вычислений**: При работе с числами с плавающей точкой важно учитывать ограничения точности, связанные с форматом IEEE 754.
- **Производительность**: Регистры SSE обычно более эффективны для современных процессоров, чем FPU.
- **Совместимость**: Убедитесь, что ваш процессор поддерживает необходимые инструкции (например, SSE).


# Ограничения точности, связанные с форматом IEEE 754

Формат IEEE 754, который используется для представления чисел с плавающей точкой в компьютерах, имеет несколько ограничений точности. Эти ограничения связаны с тем, как числа хранятся в памяти, и могут приводить к ошибкам округления, потере точности и другим проблемам. Рассмотрим основные ограничения точности, связанные с этим форматом.

## **Ограниченное количество бит для мантиссы**

Мантисса (или значимая часть числа) в формате IEEE 754 имеет фиксированное количество бит:
- **Одинарная точность (32 бита)**: 23 бита для мантиссы.
- **Двойная точность (64 бита)**: 52 бита для мантиссы.

Это означает, что не все десятичные числа могут быть точно представлены в двоичном формате. Например, число `0.1` в десятичной системе не может быть точно представлено в двоичной системе с плавающей точкой, что приводит к ошибкам округления.

### Пример:
- В одинарной точности число `0.1` представляется как приблизительно `0.10000000149011612`.
- В двойной точности число `0.1` представляется как приблизительно `0.1000000000000000055511151231257827021181583404541015625`.


## **Ошибки округления**

Из-за ограниченного количества бит в мантиссе числа округляются до ближайшего представимого значения. Это может приводить к накоплению ошибок при выполнении арифметических операций.

### Пример:
```python
a = 0.1
b = 0.2
c = a + b  # Ожидается 0.3, но результат может быть 0.30000000000000004
```

## **Потеря значимости (Underflow)**

Когда результат операции становится слишком маленьким для представления в формате IEEE 754, происходит потеря значимости. В этом случае результат округляется до нуля или денормализованного числа (очень маленького числа, близкого к нулю).

### Пример:
- В одинарной точности наименьшее положительное нормализованное число равно примерно `1.18e-38`.
- Если результат операции меньше этого значения, он может быть округлен до нуля.

## **Переполнение (Overflow)**

Когда результат операции становится слишком большим для представления в формате IEEE 754, происходит переполнение. В этом случае результат может быть округлен до бесконечности (`Inf`).

### Пример:
- В одинарной точности наибольшее представимое число равно примерно `3.4e38`.
- Если результат операции превышает это значение, он может быть округлен до `Inf`.


## **Денормализованные числа**

Денормализованные числа — это очень маленькие числа, которые не могут быть представлены в нормализованной форме из-за ограничений экспоненты. Они позволяют уменьшить потерю точности для очень маленьких чисел, но их использование может замедлять вычисления.

### Пример:
- В одинарной точности денормализованные числа находятся в диапазоне примерно от `1.4e-45` до `1.18e-38`.


## **Ошибки при сложении и вычитании**

При сложении или вычитании чисел с сильно различающимися порядками может происходить потеря точности. Это связано с тем, что младшие разряды меньшего числа теряются при выравнивании порядков.

#### Пример:
```python
a = 1e20
b = 1.0
c = a + b  # Результат может быть равен 1e20, так как 1.0 теряется на фоне 1e20
```

## **Неточность при умножении и делении**

Умножение и деление чисел с плавающей точкой также могут приводить к ошибкам округления, особенно если результат операции не может быть точно представлен в формате IEEE 754.

### Пример:
```python
a = 1.0
b = 3.0
c = a / b  # Результат 0.3333333333333333 вместо точного 1/3
```

## **Проблемы сравнения чисел**

Из-за ошибок округления сравнение чисел с плавающей точкой может быть неточным. Например, два числа, которые математически равны, могут не совпадать из-за различий в их двоичном представлении.

#### Пример:
```python
a = 0.1 + 0.2
b = 0.3
print(a == b)  # Может быть False из-за ошибок округления
```

## **Рекомендации для работы с числами с плавающей точкой**

Чтобы минимизировать проблемы, связанные с ограничениями точности, можно использовать следующие подходы:
- **Использование двойной точности**: Двойная точность (64 бита) обеспечивает большую точность, чем одинарная (32 бита).
- **Сравнение с допуском**: Вместо точного сравнения чисел используйте сравнение с допустимой погрешностью (например, `abs(a - b) < epsilon`).
- **Аккуратное выполнение операций**: Избегайте операций, которые могут привести к потере точности (например, сложение чисел с сильно различающимися порядками).

### **Числа с плавающей точкой: теоретическая база**
Числа с плавающей точкой представляются в формате IEEE 754. Этот стандарт определяет два основных типа:

- **Single precision (32 бита)**:
  - 1 бит для знака.
  - 8 бит для экспоненты.
  - 23 бита для мантиссы.

- **Double precision (64 бита)**:
  - 1 бит для знака.
  - 11 бит для экспоненты.
  - 52 бита для мантиссы.

В NASM можно использовать как single, так и double precision числа.


### **Регистры FPU (Floating Point Unit)**
Для работы с числами с плавающей точкой используется специальная FPU (Floating Point Unit), которая имеет собственный набор регистров:

- **ST(0), ST(1), ..., ST(7)** — стек из 8 регистров.
- Все операции выполняются относительно вершины стека (`ST(0)`).
- При загрузке числа оно помещается в `ST(0)`, а предыдущие значения сдвигаются.


### **Загрузка чисел с плавающей точкой**
Чтобы работать с числами с плавающей точкой, их нужно загрузить в регистры FPU. Для этого используются следующие инструкции:

#### a) Загрузка константы
```nasm
FLD real_value   ; Загружает значение real_value в ST(0)
```

#### b) Загрузка из памяти
```nasm
extern printf 
section .data
    float_num dd 3.14           ; Single precision (32-bit)
    double_num dq 2.71828       ; Double precision (64-bit)

section .text
    FLD dword [float_num]       ; Загружает float_num в ST(0)
    FLD qword [double_num]      ; Загружает double_num в ST(0)
```

#### c) Загрузка нуля
```nasm
FLDZ                ; Загружает 0.0 в ST(0)
```

#### d) Загрузка единицы
```nasm
FLD1                ; Загружает 1.0 в ST(0)
```

#### e) Загрузка π
```nasm
FLDPI               ; Загружает π (3.14159...) в ST(0)
```

### **Сохранение результата**
После выполнения вычислений результат обычно сохраняется обратно в память. Для этого используется инструкция `FSTP` (store and pop):

```nasm
section .bss
    result resd 1   ; Одиночная точность (32-bit)
    result_d resq 1 ; Двойная точность (64-bit)

section .text
    global main
main:
    mov ebp, esp; for correct debugging    
    FLD1                ; Загружает 1.0 в ST(0)
    FSTP dword [result]   ; Сохраняет ST(0) в result и очищает ST(0)
    FLD1                ; Загружает 1.0 в ST(0)
    FSTP qword [result_d] ; Сохраняет ST(0) в result_d и очищает ST(0)
    ; Выход из программы
    mov eax, eax        ; Код возврата 0
    ret
```


### **Сравнение чисел**
Для сравнения чисел с плавающей точкой используется инструкция `FCOM` или её варианты:

```nasm
FCOM st(1), st(0)   ; Сравнивает ST(1) и ST(0)
FTST                ; Сравнивает ST(0) с 0.0
FUCOM               ; Сравнение без учета NaN (Not-a-Number)
```

После сравнения флаги состояния записываются в специальный регистр FPU. Эти флаги можно проверить с помощью инструкций вроде `FSTSW AX` (сохраняет состояние в регистре AX) и последующего анализа значений.


### **Пример программы**
Простая программа на NASM, которая складывает два числа с плавающей точкой.

```nasm
extern printf\nextern printf section .data
    num1 dd 3.5         ; Первое число (single precision)
    num2 dd 4.2         ; Второе число (single precision)

section .bss
    sum resd 1          ; Переменная для хранения результата

section .text
    global main

main:
    mov ebp, esp; for correct debugging
    ; Загрузка чисел в FPU
    FLD dword [num1]    ; Загружаем num1 в ST(0)
    FADD dword [num2]   ; Складываем num1 и num2, результат в ST(0)
    ; Сохраняем результат в память
    FSTP dword [sum]    ; Сохраняем ST(0) в sum

    ; Выход из программы
    mov eax, 0        ; Код возврата 0
    ret
```

### **Особенности работы с FPU**
- **Стековая структура**: Важно помнить, что регистры FPU организованы как стек, поэтому при работе с ними нужно следить за тем, чтобы стек не переполнялся или не становился пустым.
- **Обработка исключений**: FPU может генерировать исключения (например, деление на ноль). Эти исключения можно обрабатывать с помощью специальных механизмов.
- **Точность**: При работе с числами с плавающей точкой следует учитывать ограничения точности и возможные ошибки округления.


# **ПРИМЕРЫ ПРОГРАММ**

#### **Пример 1: Загрузка и вывод числа**
**Условие:** Загрузить число `3.14` в FPU и вывести его.
- **Описание:** Программа загружает число `3.14` в регистры FPU, используя инструкцию `FLD`, сохраняет его в стеке как double precision (`qword`) и выводит через функцию `printf`.
- **Код:**
```nasm

extern printf 
section .data
    num dd 3.14         ; Число 3.14 (single precision)
    fmt db "Number: %f", 10, 0

section .text
    global main

main:
    FLD dword [num]     ; Загружаем num в ST(0)
    sub esp, 8          ; Резервируем место под double (8 байт)
    fstp qword [esp]    ; Сохраняем ST(0) в стек
    push fmt            ; Адрес формата
    call printf         ; Вызов функции printf
    add esp, 12         ; Очищаем стек
    ; Выход из программы
    xor eax, eax        ; Код возврата 0
    ret
```

---

#### **Пример 2: Сложение двух чисел**
**Условие:** Сложить два числа `2.5` и `1.7`.
- **Описание:** Программа загружает два числа в FPU, выполняет их сложение с помощью инструкции `FADD`, сохраняет результат в стеке и выводит через `printf`.
- **Код:**
```nasm
extern printf 
section .data
    num1 dd 2.5         ; Первое число
    num2 dd 1.7         ; Второе число
    fmt db "Result: %f", 10, 0

section .text
    global main

main:
    FLD dword [num1]    ; Загружаем num1 в ST(0)
    FADD dword [num2]   ; Складываем num1 и num2
    sub esp, 8          ; Резервируем место под double (8 байт)
    fstp qword [esp]    ; Сохраняем результат в стек
    push fmt            ; Адрес формата
    call printf         ; Вызов функции printf
    add esp, 12         ; Очищаем стек
    ; Выход из программы
    xor eax, eax        ; Код возврата 0
    ret
```

---

#### **Пример 3: Вычитание двух чисел**
**Условие:** Вычесть `1.7` из `2.5`.
- **Описание:** Программа выполняет вычитание с помощью инструкции `FSUB` и выводит результат.
- **Код:**
```nasm
extern printf 
section .data
    num1 dd 2.5         ; Первое число
    num2 dd 1.7         ; Второе число
    fmt db "Result: %f", 10, 0

section .text
    global main

main:
    FLD dword [num1]    ; Загружаем num1 в ST(0)
    FSUB dword [num2]   ; Вычитаем num2 из num1
    sub esp, 8          ; Резервируем место под double (8 байт)
    fstp qword [esp]    ; Сохраняем результат в стек
    push fmt            ; Адрес формата
    call printf         ; Вызов функции printf
    add esp, 12         ; Очищаем стек
    ; Выход из программы
    xor eax, eax        ; Код возврата 0
    ret
```

---

#### **Пример 4: Умножение двух чисел**
**Условие:** Умножить `3.0` на `4.0`.
- **Описание:** Программа выполняет умножение с помощью инструкции `FMUL` и выводит результат.
- **Код:**
```nasm
extern printf 
section .data
    num1 dd 3.0         ; Первое число
    num2 dd 4.0         ; Второе число
    fmt db "Result: %f", 10, 0

section .text
    global main

main:
    FLD dword [num1]    ; Загружаем num1 в ST(0)
    FMUL dword [num2]   ; Умножаем num1 на num2
    sub esp, 8          ; Резервируем место под double (8 байт)
    fstp qword [esp]    ; Сохраняем результат в стек
    push fmt            ; Адрес формата
    call printf         ; Вызов функции printf
    add esp, 12         ; Очищаем стек
    ; Выход из программы
    xor eax, eax        ; Код возврата 0
    ret
```

---

#### **Пример 5: Деление двух чисел**
**Условие:** Разделить `8.0` на `2.0`.
- **Описание:** Программа выполняет деление с помощью инструкции `FDIV` и выводит результат.
- **Код:**

#### **Пример 8: Сравнение двух чисел**
**Условие:** Сравнить `3.5` и `2.5`.
- **Описание:** Программа выполняет сравнение с помощью инструкции `FCOM` и проверяет флаги состояния с помощью `FSTSW` и `sahf`.
- **Код:**
```nasm
section .data
    num1 dd 3.5         ; Первое число
    num2 dd 2.5         ; Второе число

section .text
    global main

main:
    mov ebp, esp; for correct debugging
    FLD dword [num1]    ; Загружаем num1 в ST(0)
    FCOM dword [num2]   ; Сравниваем num1 и num2
    FSTSW ax            ; Сохраняем статус FPU в AX
    sahf                ; Передаем старшую часть AX (AH) в флаги процессора

    ; Проверяем результат сравнения
    ja greater          ; Если num1 > num2, переходим на метку greater
    jb less             ; Если num1 < num2, переходим на метку less
    jmp equal           ; Иначе числа равны

greater:
    ; Обработка случая num1 > num2
    mov eax, 1
    jmp exit

less:
    ; Обработка случая num1 < num2
    mov eax, 2
    jmp exit

equal:
    ; Обработка случая num1 == num2
    mov eax, 3

exit:
    ; Выход из программы
    xor eax, eax
    ret
```



## Работа стека при вызве функции

Стек рассчитан на косвенную адресацию через специальный регистр **ЕSP** - указатель стека (Stack Pointer). При помещении элементов в стек происходит **автоматический декремент** указателя стека, а при извлечении - **автоматический инкремент**, таким образом **стек растет в сторону младших адресов**. Адрес последнего включенного в стек элемента, содержащийся в регистре указателя стека, называется вершиной стека. Стек оперирует только **словами**, так что попытка включения байта все равно приведет к включению целого слова.

**Физический адрес стека** формируется из содержимого регистров **ЕSP** и **SS** или **ЕBP** и **SS** , причем **ЕSP** служит неявным указателем стека для всех операций включения и извлечения, а **SS** - сегментным регистром стека. Значение регистра **SS** является самым младшим адресом (т. е. границей) области стека и называется **базой стека**. Начальное значение регистра **ЕSP** есть наибольшее смещение, которого может достигать стек. Регистр **ЕBP** предназначен для произвольных обращений к стеку.

Стек эффективнее обычной памяти в двух отношениях. Команды **PUSH** и **POP** короче команды **MOV**, так как один из операндов косвенно адресуется через регистр **ЕSP** , а инкремент или декремент регистра **ЕSP** с образованием
нового целевого адреса производится автоматически.

При вызове подпрограммы необходимо выполнить следующие три действия:
- в отличие от других команд переходов вызов подпрограммы должен запомнить адрес следующей команды, чтобы можно было осуществить возврат в нужное место вызывающей программы;
- используемые подпрограммой регистры необходимо запомнить до изменения их содержимого, а перед самым выходом из подпрограммы восстановить эти регистры;
- подпрограмма должна иметь средства взаимодействия или разделения данных с вызвавшей ее программой и другими подпрограммами.

Команда **CALL** не только осуществляет переход по указанному адресу, но и помещает в стек адрес возврата (т. е. адрес команды, непосредственно следующей за командой **CALL**). Команда **RET** просто извлекает адрес возврата из стека. При этом предполагается, что если подпрограмма обращалась к стеку, то указатель стека возвращен на прежнее значение и действия со стеком не разрушили адрес возврата.

В команде **RET** допускается необязательный операнд, который прибавляется к содержимому регистра **ЕSP** после извлечения из стека адреса возврата.





Основная программа или вызывающая подпрограмма размещает в стеке параметры, затем при вызове подпрограммы в стек заносится адрес возврата, а после этого вызванная подпрограмма резервирует место в стеке под локальные переменные. Все вместе образует ***стековый фрейм***. Для корректной работы со стековым фреймом необходимо **первым действием подпрограммы** необходимо сохранить значение регистра **ESP** в регистр **EBP** и использовать его для доступа к параметрам и локальным переменным, а **ESP** продолжает работать как указатель стека. Перед **завершением подпрограммы** необходимо восстановить значение **ESP** передав в него значение из **EBP**.

Первые действия
```asm
push ebp
mov ebp, esp
sub esp, <размер локальных данных в байтах>
```

Последние действия
```asm
mov esp, ebp
pop ebp
ret
```

#### Пример

Получаем со стандартного ввода два числа, передаем их в стек, вызываем подпрограмму, которая читает числа из стека и складывает их, размещая результат в регистре **ЕАХ**. 

```asm
%include "io.inc"

section .text
global CMAIN
CMAIN:
    push ebp
    mov ebp, esp
    sub esp,8    
    GET_DEC 4,[esp]
    GET_DEC 4,[esp+4]
    call my_add
    PRINT_DEC 4,eax
    leave    
    ret
    
my_add:
    push ebp
    mov ebp, esp
    mov eax, dword [ebp+8] 
    add eax, dword [ebp+12]
    mov esp, ebp
    pop ebp
    ret
```

Команда **LEAVE** только копирует **EBP** в **ESP**, тем самым выбрасывая весь последний кадр стека, созданный последней выполненной подпрограммой, и считывает из стека **EBP** для вызывавшего кода, что одновременно восстанавливает и значение, которое имел **ESP** до вызова подпрограммы. Результатом выполнения команды **LEAVE** является корректное завершение работы состеком. 

Для обеспечения корректного начала работы со стеком существует команда **ENTER**, которая создаёт стековый кадр заданного размера и уровня вложенности (оба операнда – числа; уровень вложенности может принимать значения только от 0 до 31). 

***ВНИМАНИЕ!*** применение команды **ENTER** не рекомендуется, т.к. обоаботка параметров так и не риализована (вариант **ENTER 0, 0** всегда будет срабатывать штатно), а время ее выполнения существенно (по некоторым данным до 5 раз) превышает время выполнения стандартного пролога
```asm
push ebp
mov ebp, esp
```

Команда  **LEAVE**, наоборот, срабатывает быстрее (насколько данных нет), чем стандартый эпилог
```asm
mov esp, ebp
pop ebp
```

#### Пример

Получаем со стандартного ввода числo, передаем его в стек, вызываем подпрограмму, которая вычисляет его факториал с помощью рекурсивных вызовов, размещая результат в регистре **ЕАХ**. 
```asm
%include "io.inc"

section .text
global CMAIN
CMAIN:
    mov ebp, esp; for correct debugging
    push ebp
    mov ebp, esp
    sub esp,4    
    GET_DEC 4,[esp]
    mov ecx, dword [esp]    
    call factorial
    PRINT_DEC 4,eax
    mov esp, ebp 
    pop ebp     
    ret
    
factorial:
    cmp     ecx, 1                  ; проверяем n <= 1?
    jnbe    .rec                    ; eсли нет, делаем рекурсивный вызов
    mov     eax, 1                  ; иначе возвращаем 1
    ret
    .rec:
        push    ecx                     ; сохраняем n в стек
        dec     ecx                     ; n-1
        call    factorial               ; вызываем factorial(n-1)
        pop     ecx                     ; восстанавливаем n из стека
        imul    eax, ecx                ; n * factorial(n-1)
        ret 
```
