## Команды сдвига

Команды сдвига позволяют выполнять действия над отдельными битами операндов. Все команды сдвига перемещают биты в поле операндов влево или вправо. При выполнении команд сдвига флаг **CF** всегда содержит значение **последнего выдвинутого бита**. Рассмотрим следующие команды сдвига:

```asm
shr операнд, счетчик_сдвигов ; логический (беззнаковый) сдвиг вправо

shl операнд, счетчик_сдвигов ; логический (беззнаковый) сдвиг влево
```

При сдвигах **влево правые биты заполняются нулями**. Сдвиг **влево на один** двоичный разряд часто используется для
**удваивания** чисел, а сдвиг **на один разряд вправо** — для **деления на 2** с отбрасыванием дробной части. Эти операции осуществляются **значительно быстрее, чем команды умножения или деления**. 

#### Пример

В следующем примере команда **shr** сдвигает содержимое регистра **al** вправо на **1** бит. Выдвинутый в результате один бит попадает в флаг **CF**, а самый левый бит регистра **al** заполняется нулем.
```asm
%include "io.inc"

section .data  
x db 10110111b

section .text
global CMAIN
CMAIN:
    mov al, [x]  ; в al содержится 10110111 = b7 = 183
    shr al, 1    ; в al содержится 01011011 = 5b = 91, CF=1

    ret

```

#### Пример

```asm
%include "io.inc"

section .data  
x db 10110111b

section .text
global CMAIN
CMAIN:
    mov ax, [x]  ; в al содержится 10110111 = b7 = 183
    shl ax, 1    ; в al содержится 01011011 = 5b = 366

    ret

```

Количество позиций, на которые должен быть выполнен сдвиг, задаются вторым аргументом инструкции сдвига, и может быть либо задано **константой**, либо **регистром cl**: например:

```asm
shl ax, 12

shl eax, cl
```

#### Пример

```asm
%include "io.inc"

section .data  
x dd 0x4b7
n db 4

section .text
global CMAIN
CMAIN:
    mov ebp, esp; for correct debugging
    mov ax, [x]  ; в ax содержится 04b7 = 1207
    mov cl, [n]
    shl ax, cl   ; в ax содержится 4b70 = 2^n * 1207 = 16 * 1207 = 19312

    ret
```


#### Пример

```asm
%include "io.inc"

section .data  
x dd 0x4b70
n db 4

section .text
global CMAIN
CMAIN:
    mov ebp, esp; for correct debugging
    mov ax, [x]  ; в ax содержится 4b70 = 19312
    mov cl, [n]
    shr ax, cl   ; в ax содержится 04b7 = 19312 / 2^n = 19312 / 16 = 1207 

    ret
```

## Команды циклического сдвига

Циклический сдвиг представляет собой операцию сдвига, при которой бит, выдвинутый с одного «конца» числа, занимает освободившийся разряд с другой стороны этого же числа. Существуют следующие команды циклического сдвига:

```asm
ror ; Циклический сдвиг вправо
rol ; Циклический сдвиг влево
rcr ; Циклический сдвиг вправо с переносом
rcl ; Циклический сдвиг влево с переносом
```

В командах **rcr** и **rcl** в сдвиге участвует флаг **CF**. Выдвигаемый из регистра бит заносится в флаг **CF**, а прежнее значение **CF** при этом поступает в освободившуюся позицию в регистре.

#### Пример

```asm
%include "io.inc"

section .data  
x db 10110111b

section .text
global CMAIN
CMAIN:
    mov ax, [x]  ; в ax содержится 0000000010110111 
    ror ax, 1    ; в ax содержится 1000000001011011
                                   
    ret

```

#### Пример

```asm
%include "io.inc"

section .data  
x db 10110111b

section .text
global CMAIN
CMAIN:
    mov ax, [x]  ; в ax содержится 0000000010110111 
    rol ax, 1    ; в ax содержится 0000000101101110
                                          
    ret

```

#### Пример

```asm
%include "io.inc"

section .data  
x db 10110111b

section .text
global CMAIN
CMAIN:
    mov ax, [x]  ; в ax содержится 0000000010110111 
    rcr ax, 1    ; в ax содержится 0000000001011011 CF=1
                                   
    ret

```

#### Пример

```asm
%include "io.inc"

section .data  
x db 10110111b

section .text
global CMAIN
CMAIN:
    mov ax, [x]  ; в ax содержится 0000000010110111 
    rcl ax, 1    ; в ax содержится 0000000101101110 CF=0
                                   
    ret

```

Как уже отмечалось для **беззнаковых** чисел сдвиги соответствуют умножению или делению на $2^n$. Для **знаковых** чисел сдвиг в лево действует аналогично, а вот сдвиг в право из отрицательного числа делает положительное, т.к. обнуляет знаковый бит. Для работы с **знаковыми**  числами существуют команды **арифметического сдвига**

```asm
sar операнд, счетчик_сдвигов ; знаковый сдвиг вправо

sal операнд, счетчик_сдвигов ; знаковый сдвиг влево
```

## Способы адресации

### Прямая адресация

Прямая адресация - это адресация по смещению. Если известен адрес операнда, располагающегося в памяти, можно использовать этот адрес. Если операнд – слово, находящееся в сегменте, на который указывает **ES**, со смещением от начала сегмента **0001**, то команда

```asm
mov ax,es:0001
```

поместит это слово в регистр **AX**. 

При использовании для создания статических переменных директив определения данных в разделе **.data**, которые позволяют ссылаться на статические переменные не по адресу, а по имени. В этом случае, если в сегменте, указанном в **ES**, была описана переменная **х** размером в слово, можно записать ту же команду как

```asm
mov ax,es:х
```

Если селектор сегмента данных находится в DS, имя сегментного регистра при прямой адресации можно не указывать, т.к. DS используется по умолчанию. 

### Косвенная адресация

По аналогии с регистровыми и непосредственными операндами адрес операнда в памяти также можно не указывать непосредственно, а хранить в любом регистре. До 80386 для этого можно было использовать только BX, SI, DI и BP, но потом эти ограничения были сняты и в настоящее время адрес операнда разрешили считывать также и из **EAX, EBX, ECX, EDX, ESI, EDI, EBP и ESP** (но не из AX, CX, DX или SP напрямую – надо использовать EAX, ECX, EDX, ESP соответственно или предварительно скопировать смещение в BX, SI, DI или BP). Например, следующая команда помещает в регистр **AX** слово из ячейки памяти, селектор сегмента которой находится в **DS**, а смещение – в **BX**:

```asm
mov ax,[bx]
```

Как и в случае прямой адресации, DS используется по умолчанию, но не во всех случаях: если смещение берут из регистров ESP, EBP или BP, то в качестве сегментного регистра используется SS. 

### Адресация по базе со сдвигом

Cледующая команда

```asm
mov ax,[bx+2]
```

помещает в регистр AX слово, находящееся в сегменте, указанном в DS, со смещением на 2 большим, чем число, находящееся в BX. Так как слово занимает ровно два байта, эта команда поместила в AX слово, непосредственно следующее за тем, которое есть в предыдущем примере. 

Эта форма адресации используется в тех случаях, когда в регистре находится адрес начала структуры данных (например массива), а доступ надо осуществить к какому-либо элементу этой структуры. 

Другие допустимые формы записи этого способа адресации:

```asm
mov ax,[bp]+2
mov ax,2[bp]
```

С помощью этого метода можно организовывать доступ к одномерным массивам байт: смещение соответствует адресу начала массива, а число в регистре – индексу элемента массива, который надо считать. 

Если массив состоит не из байт, а из
- слов, необходимо умножать базовый регистр на два, 
- двойных слов – на четыре. 

Для этого предусмотрен следующий специальный метод адресации.

### Косвенная адресация с масштабированием

Этот метод адресации полностью идентичен предыдущему, за исключением того, что с его помощью можно прочитать элемент массива слов,двойных слов или учетверенных слов, просто поместив номер элемента
в регистр:

```asm
mov ax,[esi*2]+2
```

Множитель, который может быть равен 1, 2, 4 или 8, соответствует размеру элемента массива – байту, слову, двойному слову, учетверенному слову.

При таком варианте адресации можно использовать только регистры **EAX, EBX, ECX, EDX, ESI, EDI, EBP, ESP**, но не
SI, DI, BP или SP, которые можно было использовать в предыдущих вариантах.

### Адресация по базе c индексированием

В этом методе адресации смещение операнда в памяти вычисляется как сумма чисел, содержащихся в двух регистрах, и смещения, если оно указано. Все следующие команды – это разные формы записи одного и того же действия:

```asm
mov ax,[bx+si+2]
mov ax,[bx][si]+2
mov ax,[bx+2][si]
mov ax,[bx][si+2]
mov ax,2[bx][si]
```

В регистр **AX** помещается слово из ячейки памяти со смещением, равным сумме чисел, содержащихся в **BX, SI**, и числа 2. 

Из шестнадцатибитных регистров так можно складывать только BX + SI, BX + DI, BP + SI и BP + DI, а из 32-битных – все восемь регистров общего назначения. 

Так же как и для прямой адресации, вместо непосредственного указания числа можно использовать имя переменной, заданной одной из директив определения данных. 

## Описание и инициализация массива в программе

Специальных средств описания массивов в языке ассемблера **НЕТ**. При необходимости использовать массив в программе его нужно моделировать одним из следующих способов:

- Перечислением элементов массива в поле операндов одной из директив описания данных. При перечислении элементы разделяются запятыми
```asm
arr dd 1,2,3,4,5 ;массив из 5 элементов.Размер каждого элемента 4 байта
```

- Используя оператор повторения **times**
```asm
arr times 5 dw 0 ;массив из 5 нулевых элементов. Размер каждого элемента 2 байта
```

Так можно считать, например, число из двумерного массива.

Если массивы состоят не из одиночных байт, а из слов или двойных слов, удобнее использовать следующую, наиболее полную форму адресации.

### Адресация по базе с индексированием и масштабированием

Это самая полная возможная схема адресации, в которую входят все случаи, рассмотренные ранее, как частные. Полный адрес операнда можно записать как выражение, представленное на рисунке

## Доступ к элементам массива

При работе с массивами необходимо учитывать, что **все элементы массива располагаются в памяти компьютера последовательно**. Само по себе такое расположение ничего не говорит о назначении и порядке использования этих элементов. И только программист с помощью составленного им алгоритма обработки определяет, как нужно трактовать эту последовательность байт, составляющих массив. 

**Одну и ту же область памяти** можно трактовать **как одномерный массив**, и одновременно те же самые данные могут трактоваться **как двухмерный массив**. 

Эти же соображения можно распространить и на индексы элементов массива. Ассемблер не подозревает об их существовании и ему абсолютно все равно, каковы их численные смысловые значения. Для того чтобы локализовать определенный элемент массива, к его имени нужно добавить индекс. Так как мы моделируем массив, то должны позаботиться и о моделировании индекса. В языке ассемблера индексы массивов — это обычные адреса, но с ними работают особым образом. Другими словами, когда **при программировании на ассемблере мы говорим об индексе, то скорее подразумеваем под этим не номер элемента в массиве, а некоторый адрес**.

#### Пример

Создать нулевую матрицу размера 5х5. Считать ввод номеров строки и столбца элемента, которому присвоить значение 5.

```asm
%include "io.inc"

section .bss
i resd 1
j resd 1

n equ 5
m equ 5  
arr resw (n*m) 

section .text
global CMAIN
CMAIN:
    mov ebp, esp; for correct debugging
    
    GET_DEC 2, [i]
    GET_DEC 2, [j]
    PRINT_DEC 2, [i]
    NEWLINE
    PRINT_DEC 2, [j]

    mov esi, [i]                  ; индекс i             
    mov edi, [j]                  ; индекс j    

    imul ecx, esi, n              ; смещаемся на esi строк по n элементов    
    add  ecx, edi                 ; и еще на edi элементов в строке
    
    mov word [arr+2*ecx], 5       ; записываем значение
    
    ret
```
